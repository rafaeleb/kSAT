## Parameter Plots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def plotter(k,nvar,p):
    df = pd.read_csv("data/pars_k_"+str(k)+"_nvar_"+str(nvar)+"_layers_"+str(p)+".csv")
    df = df.apply(lambda x: abs(x))
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    mean_df = df.mean()
    stdv_df = df.std()
    lm_y = [mean_df[i] for i in range(0,2*p)]
    lm_x = [i+1 for i in range(0,2*p)]
    lm_s = [stdv_df[i] for i in range(0,2*p)]
    return (lm_x, lm_y, lm_s)
```

```python
p1 = plotter(3,3,2)
p2 = plotter(3,3,3)
p3 = plotter(3,5,2)
p4 = plotter(3,5,3)
p5 = plotter(3,7,2)
p6 = plotter(3,7,3)
p7 = plotter(5,5,3)
```

```python
fig = plt.figure()
plt.figure(figsize=(6,4))

plt.axhspan(xmin = 0, xmax = 0.32, ymin = 0.1, ymax = 1, facecolor='gray', alpha=0.2)
plt.axhspan(xmin = 0.32, xmax = 0.69, ymin = 0.1, ymax = 1, facecolor='gray', alpha=0.4)
plt.axhspan(xmin = 0.69, xmax = 1, ymin = 0.1, ymax = 1, facecolor='gray', alpha=0.6)

plt.errorbar(p1[0], p1[1], p1[2], marker = 'o', label = 'N=3, p=2')
plt.errorbar(p2[0], p2[1], p2[2], marker = 'o', label = 'N=3, p=3')
plt.errorbar(p3[0], p3[1], p3[2], marker = 'o', label = 'N=5, p=2')
plt.errorbar(p4[0], p4[1], p4[2], marker = 'o', label = 'N=5, p=3')
plt.errorbar(p5[0], p5[1], p5[2], marker = 'o', label = 'N=7, p=2')
plt.errorbar(p6[0], p6[1], p6[2], marker = 'o', label = 'N=7, p=3')

#plt.yscale('log')
plt.xlabel("Parameter number")
plt.ylabel("Parameter absolute value")
plt.title('a) k=3', loc='left')
plt.grid(c = 'black')
plt.legend(loc='lower right')
plt.savefig('figs/params_k3.png')
```

```python
p553 = plotter(5,5,3)
p573 = plotter(5,7,3)
p593 = plotter(5,9,3)
p554 = plotter(5,5,4)
p574 = plotter(5,7,4)
```
