```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plotter(k,nvar,p):
    df = pd.read_csv("data/pars_k_"+str(k)+"_nvar_"+str(nvar)+"_layers_"+str(p)+".csv")
    df = df.apply(lambda x: abs(x))
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    mean_df = df.mean()
    lm_y = [mean_df[i] for i in range(0,2*p)]
    lm_x = [i+1 for i in range(0,2*p)]
    return (lm_x, lm_y)
```

```python
p1 = plotter(3,3,2)
p2 = plotter(3,3,3)
p3 = plotter(3,5,2)
p4 = plotter(3,5,3)
p5 = plotter(3,7,2)
p6 = plotter(3,7,3)
```

```python
plt.axhspan(xmin = 0, xmax = 0.32, ymin = 0.1, ymax = 1, facecolor='gray', alpha=0.2)
plt.axhspan(xmin = 0.32, xmax = 0.69, ymin = 0.1, ymax = 1, facecolor='gray', alpha=0.4)
plt.axhspan(xmin = 0.69, xmax = 1, ymin = 0.1, ymax = 1, facecolor='gray', alpha=0.6)

plt.scatter(p1[0], p1[1], label="N=3, p=2")
plt.plot(p1[0], p1[1])

plt.scatter(p2[0], p2[1], label="N=3, p=3")
plt.plot(p2[0], p2[1])

plt.scatter(p3[0], p3[1], label="N=5, p=2")
plt.plot(p3[0], p3[1])

plt.scatter(p4[0], p4[1], label="N=5, p=3")
plt.plot(p4[0], p4[1])

plt.scatter(p5[0], p5[1], label="N=7, p=2")
plt.plot(p5[0], p5[1])

plt.scatter(p6[0], p6[1], label="N=7, p=3")
plt.plot(p6[0], p6[1])

plt.xlabel("Parameter number")
plt.ylabel("Paramter absolute value")

plt.rc('axes', axisbelow=True)
plt.grid()

plt.legend(loc="lower right")
```
