```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
t_tab = []
s_tab = []

for i in range(4, 10):
    df = pd.read_csv('times_k_4_vars_'+str(i)+'_p_5.csv')
    t_mean = df['training'].mean()
    s_mean = df['sampling'].mean()
    t_tab.append(t_mean)
    s_tab.append(s_mean)
```

```python
plt.figure(figsize=(6,2))
plt.plot([i for i in range(4,10)],t_tab, linestyle = 'dashed', marker = 'o', color = "orange")
plt.xlabel("Number of variables")
plt.ylabel("Training time (s)")
plt.title('Training time 4-SAT, p=5 layers, 2048 samples')
plt.show()
```

```python
plt.figure(figsize=(6,2))
plt.plot([i for i in range(4,10)], s_tab, linestyle = 'dashed', marker = 'o', color = "firebrick")
plt.xlabel("Number of variables")
plt.ylabel("Sampling time (s)")
plt.title('Sampling time 4-SAT, p=5 layers, 2048 samples')
plt.show()
```
